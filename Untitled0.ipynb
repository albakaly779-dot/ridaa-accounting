```python
import torch
from diffusers import DiffusionPipeline, DPMSolverMultistepScheduler
from diffusers.utils import export_to_video
from IPython.display import Video
import matplotlib.pyplot as plt

print("جاري تحميل نموذج الذكاء الاصطناعي... (اصبر قليلاً)")

# 2. تحميل النموذج
pipe = DiffusionPipeline.from_pretrained(
    "damo-vilab/text-to-video-ms-1.7b",
    torch_dtype=torch.float16,
    variant="fp16"
)
pipe.scheduler = DPMSolverMultistepScheduler.from_config(pipe.scheduler.config)
# pipe.enable_model_cpu_offload() # Removed this line as it requires a GPU accelerator

# 3. دالة الإنشاء
def create_video_on_phone(prompt):
    print(f"جاري العمل على: {prompt}")
    video_frames = pipe(prompt, num_inference_steps=25).frames[0]
    output_path = "mobile_video.mp4"
    export_to_video(video_frames, output_path)
    return output_path

# 4. التنفيذ
# غيّر النص هنا لما تريد (باللغة الإنجليزية)
my_prompt = "A robot dancing in the rain, cyberpunk city, 4k quality"

video_file = create_video_on_phone(my_prompt)

print("تم الانتهاء! الفيديو بالأسفل:")
# عرض الفيديو مباشرة في المتصفح
Video(video_file, embed=True, width=400)
```
